```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargamos el Data Frame
df= pd.read_csv("Tokyo.csv")
df
```

```python
#Filtramos 5 host name diferentes FILTRAMOS LA VARIBALE "host_name"
filtro1= df[(df["host_name"] == "Yoshimi & Marek") | (df["host_name"] == "Yoshimi & Marek") | (df["host_name"] == "Kei") | (df["host_name"] == "Best Stay In Tokyo!") | (df["host_name"] == "Wakana") ] 
filtro1
filtro1.to_csv("host_name.csv")
```

```python
#Host que se hallan unido antes del año 2019, FILTRAMOS LA VARIABLE "host_since"
filtro2=df[df["host_since"] < "2020-01-01"]
filtro2
filtro2.to_csv("host_since.csv")
```

```python
#Host que responden a más tardar en 1 dias, FILTRAMOS LA VARIABLE "host_response_time"
filtro3= df[(df["host_response_time"] == "within a day")] 
filtro3
filtro3.to_csv("host_response_time.csv")
```

```python
#Registros que tengan al menos 5 accommodates, FILTRAMOS LA VARIABLE "accommodates"
filtro4=df[df["accommodates"] >= 5]
filtro4
filtro4.to_csv("accommodates.csv")
```

```python
#Registros con tipo de cuarto en Entire home y Private room, FILTRAMOS LA VARIABLE "room_type"
filtro5=df[(df["room_type"] == "Private room") | (df["room_type"] == "Entire home/apt")]
filtro5
filtro5.to_csv("room_type.csv")
```

```python
#Registros que a lo mucho cuenten con 4 camas, FILTRAMOS LA VARIABLE "beds"
filtro6=df[df["beds"] <= 4]
filtro6
filtro6.to_csv("beds.csv")
```

```python
#Registros que tengan 1 baño O que sea superhost PERO con la condicion de que sean instant_bookable
#FILTRAMOS LAS VARIABLES "bathrooms","host_is_superhost","instant_bookable"
filtro7=df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtro7
filtro7.to_csv("bathroomsOhost_is_superhostYinstant_bookable.csv")
```

```python
#Registros cuyo review_scores_cleanliness sea menor a 3 Y review_scores_rating sea mayor a 3
#FILTRAMOS LAS VARIABLES "review_scores_cleanliness" Y review_scores_rating
filtro8=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
filtro8.to_csv("review_scores_cleanliness&review_scores_rating.csv")
```
